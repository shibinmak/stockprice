**PART 1 DATA PREPROCESSING**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
# Importing the training set
data = pd.read_excel('AMZN.xlsx')
data.head()
```

```python
training_set = data[['Close']].values
type(training_set)
```

```python
# Feature Scaling
sc = MinMaxScaler()
training_set = sc.fit_transform(training_set)

# Getting the inputs and the ouputs
X_train = training_set[0:-1]
y_train = training_set[1:]

# Reshaping for keras acceptable dimension
X_train = np.reshape(X_train, (len(X_train), 1, 1))

```

**PART 2 DEEP LEARNING**

```python
# Importing the Keras libraries and packages & #Building the RNN
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM

# Initialising the RNN
model = Sequential()

# Adding the input layer and the LSTM layer
# input_shape=(None, 1) = input_shape=(timestep_count, features_count)
model.add(LSTM(units=4, activation='sigmoid', input_shape=(None, 1)))

# Adding the output layer
model.add(Dense(units=1))

# Compiling the RNN
model.compile(optimizer='adam', loss='mean_squared_error')

# Fitting the RNN to the Training set
model.fit(X_train, y_train, batch_size=5, epochs=500)


#model.summary() to get a summary of the NN


```

**PART 3 PREDICTION**

```python
data.iloc[-1] #last known detail
```

```python
#as we trained our model with timestep 1, we have to find next day closing price from last day closing price.
a=data[['Close']].iloc[-1][0] #last known closing price
#range(14) because nov 21 to dec 4 = 14 days
for i in range(14):
    inputs = a    
    inputs = sc.transform(inputs)
    inputs = np.reshape(inputs, (len(inputs), 1, 1))
    predicted_stock_price = model.predict(inputs)
    predicted_stock_price = sc.inverse_transform(predicted_stock_price)
    a=predicted_stock_price[0][0] #update last known closing price
    
dec_five = a

print ("DEC 5 PREDICTED CLOSING PRICE OF AMAZON  {:.3f}".format(dec_five)) #python3
#print "DEC 5 PREDICTED CLOSING PRICE OF AMAZON  {:.3f}".format(dec_five) #python2
        
        
```
